```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import statistics
```

```python
c = int(input())
```

```python
l = []
for i in range(c):
  l.append(int(input()))
```

```python
m1 = l[4]
m2 = l[1]
print(m1)
print(m2)
```

```python
c1 = []
c2 = []
d1 = []
d2 = []
c3 = []
c4 = []
```

```python
c = len(l)
c1 = []
c2 = []
c3 = []
c4 = []

while True:
    d1 = []
    d2 = []
    for i in range(c):
        d1.append(abs(l[i] - m1))
        d2.append(abs(l[i] - m2))

    print("Difference:")
    for i in range(c):
        if d1[i] < d2[i]:
            c1.append(abs(l[i]))
        else:
            c2.append(abs(l[i]))
    print("Cluster:")
    print(c1)
    print(c2)

    if c1 == c3 and c2 == c4:
        print("Cluster: ", c1, ' ', c2, ' ')
        break
    else:
        c3 = c1
        c4 = c2
        print("Cluster:")
        print(c3)
        print(c4)
        m1 = statistics.mean(c1)
        m2 = statistics.mean(c2)
        print("Means are:")
        print(m1)
        print(m2)
```
