```python
import numpy as np
```

```python
I1 = np.random.choice([0,1])
W1 = np.random.choice([-1,1])
I2 = np.random.choice([0,1])
W2 = np.random.choice([-1,1])
T = 2
#T values:
#AND=2, OR=1, NAND=-2, NOR=0
```

```python
print (I1,W1,I2,W2)
```

```python
a = I1*W1 + I2*W2
print(a)
```

```python
if(a >= T):
  print(I1,'  ',I2,'  ',W1,'  ',W2,'  ',T,'  ',1)
else:
  print(I1,'  ',I2,'  ',W1,'  ',W2,'  ',T,'  ',0)
```
