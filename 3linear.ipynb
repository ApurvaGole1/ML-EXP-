```python
import numpy as np
```

```python
a = np.array([[1,2,3,4,5],[34,35,67,89,90]])
xmean = a[0].mean()
ymean = a[1].mean()
```

```python
A = a[0] - xmean
B = a[1] - ymean
```

```python
AB = A*B
A2 = A**2
print(AB)
print(A2)
```

```python
sumab = sum(AB)
suma2 = sum(A2)
```

```python
beta_1 = sumab/suma2
beta_0 = ymean - beta_1*xmean
```

```python
print(beta_1)
print(beta_0)
```

```python
print("y = ", beta_0,"+",beta_1," * x ")
```
